## Exercise `ex2`

Write a function `reverse_list(lst)` that returns a new list with the elements of `lst` in reverse order.

**Expected time complexity:** O(n)

```python
%%notebook_ta ex2
def reverse_list(lst):
    l = len(lst)
    for i in range(l - 1, -1, -1):
        lst[i], lst[l - 1 - i] = lst[l - 1 - i], lst[i]
    return lst
```
